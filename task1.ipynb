```python
import numpy as np
import pandas as pd
```

# load dataset

```python
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df.info()
#info function returns the summary of the dataset, including the number of rows and columns, column names ,datatype etc
```

```python
#understand the dataset, how many columns and rows are present inthe dataset.
df.shape
# in this dataset has 150 rows and  4 columns
```

```python
df.columns
#columns  function returns the column names
```

```python
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
```

# Perform data cleaning:
    -Handle missing values

```python
df.isna().sum()
```
